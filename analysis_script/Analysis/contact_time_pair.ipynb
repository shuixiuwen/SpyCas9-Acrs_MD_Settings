```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
```

```python
data_5xbl = pd.read_table("5xbl_5A_contact_out_compare.dat", sep= "\s+")
data_5xbl_rep1 = pd.read_table("5xbl_5A_contact_out_compare_rep1.dat", sep= "\s+")
data_5xbl_rep2 = pd.read_table("5xbl_5A_contact_out_compare_rep2.dat", sep= "\s+")
data_6ifo = pd.read_table("6ifo_5A_contact_out_compare.dat", sep= "\s+")
data_6ifo_rep1 = pd.read_table("6ifo_5A_contact_out_compare_rep1.dat", sep= "\s+")
data_6ifo_rep2 = pd.read_table("6ifo_5A_contact_out_compare_rep2.dat", sep= "\s+")
```

```python
fig, ((ax1,ax2),(ax3, ax4)) = plt.subplots(nrows=2, ncols=2, 
                                           dpi=300,figsize=(30, 15))

ax1.plot( data_5xbl.iloc[:,1][::10], color = '#1f77b4', 
         linewidth = 1.5, alpha = 0.70)
ax1.plot( data_5xbl_rep1.iloc[:,1][::10], color = '#ff7f0e', 
         linewidth = 1.5, alpha = 0.70)
ax1.plot( data_5xbl_rep2.iloc[:,1][::10], color = '#2ca02c', 
         linewidth = 1.5, alpha = 0.70)

ax2.plot( data_5xbl.iloc[:,2][::10], color = '#1f77b4', 
         linewidth = 1.5, alpha = 0.70, label = 'MD_1')
ax2.plot( data_5xbl_rep1.iloc[:,2][::10], color = '#ff7f0e', 
         linewidth = 1.5, alpha = 0.70, label = 'MD_2')
ax2.plot( data_5xbl_rep2.iloc[:,2][::10], color = '#2ca02c', 
         linewidth = 1.5, alpha = 0.70, label = 'MD_3')

ax2.legend(ncol = 1, bbox_to_anchor=(1, 0.5), loc='lower left', 
           borderaxespad=0.2, frameon = True, fontsize = 30, 
           numpoints = 20, markerscale = 10)

ax3.plot( data_6ifo.iloc[:,1][::10], color = '#1f77b4', 
         linewidth = 1.5, alpha = 0.70)
ax3.plot( data_6ifo_rep1.iloc[:,1][::10], color = '#ff7f0e', 
         linewidth = 1.5, alpha = 0.70)
ax3.plot( data_6ifo_rep2.iloc[:,1][::10], color = '#2ca02c', 
         linewidth = 1.5, alpha = 0.70)

ax4.plot( data_6ifo.iloc[:,2][::10], color = '#1f77b4', 
         linewidth = 1.5, alpha = 0.70)
ax4.plot( data_6ifo_rep1.iloc[:,2][::10], color = '#ff7f0e', 
         linewidth = 1.5, alpha = 0.70)
ax4.plot( data_6ifo_rep2.iloc[:,2][::10], color = '#2ca02c', 
         linewidth = 1.5, alpha = 0.70)

ax1.set_ylabel("Contact numbers", fontsize = 30)
ax3.set_ylabel("Contact numbers", fontsize = 30)
for ax in [ax1, ax2, ax3, ax4]:
    #ax.set_ylabel("H-bonds", fontsize = 30)
    ax.tick_params(width=2, labelsize=30)
    ax.set_xlim(0, len(data_5xbl))
    bwith = 1.5
    ax.spines['bottom'].set_linewidth(bwith)
    ax.spines['left'].set_linewidth(bwith)
    ax.spines['top'].set_linewidth(bwith)
    ax.spines['right'].set_linewidth(bwith)
    ax.grid(axis = 'y', ls = ':', color = "gray", alpha = 0.9)

ax3.set_xlabel("Time (ns)", fontsize = 30)
ax4.set_xlabel("Time (ns)", fontsize = 30)
ax3.set_xticks([20000, 40000, 60000, 80000, 100000])
ax3.set_xticklabels([200, 400, 600, 800, 1000])
ax4.set_xticks([20000, 40000, 60000, 80000, 100000])
ax4.set_xticklabels([200, 400, 600, 800, 1000])

for ax in [ax2, ax1]:
    ax.set(xticklabels=[])
# ax2.set_ylim(0, 200)
# ax4.set_ylim(0, 200)

fig.subplots_adjust(hspace = 0.15, wspace = 0.15)
plt.savefig('contact_num_pair_2x2.tif', dpi=300, bbox_inches='tight')
plt.show()
```
